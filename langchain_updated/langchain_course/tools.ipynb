  <p style="font-size: 15px; color: #3c4043; margin-top: 8px;">Previously, this notebook performed a standard chat query (<code>"Why do parrots talk?"</code>) which did not utilize any tools, making the output just standard text rather than tool-calling data. This updated notebook demonstrates how to define custom tools, bind them to the model, and trigger structured tool calls.</p>
</div>

<div style="font-family: 'Segoe UI', Arial, sans-serif; line-height: 1.6; margin-top: 10px;">
  <h2 style="color: #202124; font-size: 20px; font-weight: 500; margin-bottom: 10px;">1. Setup Environment Variables and Load Model</h2>
  <p style="font-size: 14px; color: #5f6368;">Load your Groq API credentials and initialize the <code>llama-3.3-70b-versatile</code> model.</p>
</div>

```python
import os
from dotenv import load_dotenv
from langchain.chat_models import init_chat_model

# Load environment variables from .env file
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

# Initialize model
model = init_chat_model("groq:llama-3.3-70b-versatile", temperature=0)
```

<div style="font-family: 'Segoe UI', Arial, sans-serif; line-height: 1.6; margin-top: 10px;">
  <h2 style="color: #202124; font-size: 20px; font-weight: 500; margin-bottom: 10px;">2. Define a Custom Tool</h2>
  <p style="font-size: 14px; color: #5f6368;">We use LangChain's <code>@tool</code> decorator to define a simple multiplication function. The docstring and parameter typing are used by the model to understand when and how to call this tool.</p>
</div>

```python
from langchain_core.tools import tool

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two integers together.

    Args:
        a: First integer
        b: Second integer
    """
    return a * b
```

<div style="font-family: 'Segoe UI', Arial, sans-serif; line-height: 1.6; margin-top: 10px;">
  <h2 style="color: #202124; font-size: 20px; font-weight: 500; margin-bottom: 10px;">3. Bind the Tool to the Model</h2>
  <p style="font-size: 14px; color: #5f6368;">Use the <code>bind_tools</code> method to link the tool to our model instance.</p>
</div>

```python
model_with_tools = model.bind_tools([multiply])
```

<div style="font-family: 'Segoe UI', Arial, sans-serif; line-height: 1.6; margin-top: 10px;">
  <h2 style="color: #202124; font-size: 20px; font-weight: 500; margin-bottom: 10px;">4. Invoke the Model and Verify the Tool Call Output</h2>
  <p style="font-size: 14px; color: #5f6368;">Ask a question requiring math/multiplication. The model will identify that the <code>multiply</code> tool should be called, and the output will contain a structured tool call dictionary rather than just regular text.</p>
</div>

```python
response = model_with_tools.invoke("What is 123 multiplied by 456?")
print("Generated Tool Calls:")
print(response.tool_calls)

print("\nFull AIMessage Object:")
response
```
