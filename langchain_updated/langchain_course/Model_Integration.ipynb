# Model Integration with Gemini, Groq, and OpenAI

This notebook demonstrates how to initialize and invoke chat models from different providers (Gemini, Groq, and OpenAI) using LangChain's unified `init_chat_model` function.

## 1. Setup Environment Variables
Load credentials from the `.env` file.

```python
import os
from dotenv import load_dotenv
load_dotenv()

# Set environment variables for LangChain providers
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

## 2. Google Gemini Integration (Working)
Initialize and call Google's `gemini-2.5-flash` model.

```python
from langchain.chat_models import init_chat_model

print("Initializing Gemini model...")
gemini_model = init_chat_model("gemini-2.5-flash", model_provider="google_genai")

response = gemini_model.invoke("Hello, how are you?")
print("Gemini Response:")
print(response.content)
```

USING CHAT_MODEL IMPORT INVOKE NOT INIT CHAT MODELS

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
response=model.invoke("What is the sky color?")
response
```

## 3. Groq Integration (Working)
Initialize and call Groq's `llama-3.3-70b-versatile` model.

```python
print("Initializing Groq model...")
groq_model = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")

response = groq_model.invoke("Hello, how are you?")
print("Groq Response:")
print(response.content)
```
