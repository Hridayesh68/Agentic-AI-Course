  <p style="font-size: 14px; color: #5f6368;">Load your API keys and credentials from the <code>.env</code> file into the environment.</p>

```python
import os
from dotenv import load_dotenv
load_dotenv()

# Set environment variables for LangChain providers
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

<div style="font-family: 'Segoe UI', Arial, sans-serif; line-height: 1.6; margin-top: 10px;">
  <h2 style="color: #1a73e8; font-size: 20px; font-weight: 500; margin-bottom: 10px;">2. Google Gemini Integration <span style="font-size: 12px; font-weight: normal; color: #34a853; background-color: #e6f4ea; padding: 2px 8px; border-radius: 4px; margin-left: 8px;">Active / Working</span></h2>
  <p style="font-size: 14px; color: #5f6368;">Initialize and call Google's <code>gemini-2.5-flash</code> model using the <code>google_genai</code> integration package.</p>
</div>

```python
from langchain.chat_models import init_chat_model

print("Initializing Gemini model...")
gemini_model = init_chat_model("gemini-2.5-flash", model_provider="google_genai")

response = gemini_model.invoke("Hello, how are you?")
print("Gemini Response:")
print(response.content)
```

USING CHAT_MODEL IMPORT INVOKE NOT INIT CHAT MODELS

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model=ChatGoogleGenerativeAI(model="gemini-2.5-flash")
response=model.invoke("What is the sky color?")
response
```

```python
print("Initializing Groq model...")
groq_model = init_chat_model("llama-3.3-70b-versatile", model_provider="groq")

response = groq_model.invoke("Hello, how are you?")
print("Groq Response:")
print(response.content)
```

USING CHAT_MODEL IMPORT INVOKE NOT INIT CHAT MODELS FOR GROQ

```python
import os 
from langchain_groq import ChatGroq
model=ChatGroq(model="qwen/qwen3-32b")
response=model.invoke("Why do parrots talk")
response
```

Streaming

```python
for chunk in model.stream("which model are you also write a 200 words paragraph on AI"):
    print(chunk.text,end="|",flush=True)
```

Batch

```python
responses=model.batch([
    "Why do parrots have colorful feathers?",
    "How do airplanes fly?",
    "What is spacex"
],
    config={
        'max_concurreny':5,
    }

)
responses
```
